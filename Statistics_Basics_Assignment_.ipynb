1.Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss
nominal, ordinal, interval, and ratio scales.

.
Data can be broadly categorized into two types: qualitative (categorical) and quantitative (numerical). Each type serves a different purpose in data analysis.

. Qualitative Data (Categorical Data) :-
Qualitative data describes characteristics or qualities that cannot be measured numerically. Instead, they are classified into categories. This type of data is typically represented in words, labels, or symbols.

.Examples of Qualitative Data:

Eye color (blue, green, brown)

Marital status (single, married, divorced)

. Quantitative Data (Numerical Data)
Quantitative data consists of numerical values that can be measured or counted. It can be further divided into two types:

Discrete Data:- Countable values (e.g., the number of students in a class).

Continuous Data: Measurable values that can take any value within a range (e.g., height, weight, temperature).

.Examples of Quantitative Data:

Age of a person (25 years)

Weight of an object (68.5 kg)

.Nominal Scale (Qualitative):
The simplest level of measurement.

Data is categorized without any order or ranking.

Only labels or names are assigned to different groups.

.Ordinal Scale (Qualitative)
Data is categorized with a meaningful order or ranking.

The differences between categories are not uniform or measurable.

.Interval Scale (Quantitative)

Numeric scale with equal intervals between values.

No true zero point, meaning zero does not indicate the total absence of the property.

.Ratio Scale (Quantitative)

The highest level of measurement.

Has a true zero point, meaning zero represents the total absence of the measured property.

Allows for meaningful comparisons, including ratios (e.g., "twice as much").

2.What are the measures of central tendency, and when should you use each? Discuss the mean, median,
and mode with examples and situations where each is appropriate.

.Measures of central tendency summarize a dataset by identifying a single value that represents the center or typical value. The three main measures are mean, median, and mode.

.Mean (Arithmetic Average)

The mean is the sum of all values divided by the number of values in the dataset.

. Best for symmetrical distributions without extreme values.

Useful in financial analysis, performance metrics, and scientific research.

. Median (Middle Value)

The median is the middle number when the data is ordered from smallest to largest. If the dataset has an even number of values, the median is the average of the two middle values.

.Example Situation:

Income data: If salaries in a company are $30,000, $35,000, $40,000, $45,000, and $1,000,000, the mean would be misleading, but the median ($40,000) better represents typical income.

.Mode (Most Frequent Value)

The mode is the number that appears most frequently in a dataset. A dataset can have:

One mode (Unimodal)

Two modes (Bimodal)

Multiple modes (Multimodal)

No mode (if all values are unique)

.Example:

For the dataset 3, 7, 7, 10, 15, 7, 20, the mode is 7 (since it appears most often).

For the dataset 1, 2, 2, 3, 3, 4, 5, the modes are 2 and 3 (Bimodal).

5.Discuss the role of random sampling in making inferences about populations.

.Role of Random Sampling in Making Inferences About Populations

.What Is Random Sampling?

Random sampling is a statistical method used to select a representative subset of a population to draw conclusions about the entire population. Each member of the population has an equal chance of being chosen, reducing bias and ensuring the sample reflects the population‚Äôs characteristics.

.Why Is Random Sampling Important?

Ensures Representativeness: A well-selected random sample accurately reflects the diversity of the population.

Reduces Bias: Eliminates systematic favoritism in selecting individuals.
Enables Generalization: Results from a sample can be extrapolated to the whole population.

Improves Efficiency: Studying a subset is faster and more cost-effective than analyzing the entire population.

6.Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

.Concept of Skewness

Skewness refers to the asymmetry of a dataset‚Äôs distribution around its mean. A dataset is symmetrical if the left and right sides of its distribution are mirror images. If the distribution is not symmetrical, it is skewed.

Effects of Skewness on Data Interpretation

. Choosing the Right Measure of Central Tendency

Symmetrical data: The mean is the best measure of central tendency.

Skewed data: The median is a better representative measure because the mean is influenced by extreme values.

.Impact on Standard Deviation and Variance

A highly skewed distribution increases variance and standard deviation, making them less reliable as measures of spread.

11.Explain what a random variable is and differentiate between discrete and continuous random variables.

.A random variable is a numerical outcome of a random event or experiment. It is a function that assigns a real number to each possible outcome of a random phenomenon. Random variables are used to quantify the results of probabilistic experiments and are key elements in probability theory and statistics.

.There are two main types of random variables:

Discrete Random Variables

Continuous Random Variables

. Discrete Random Variables

A discrete random variable can take on countable values. These values are distinct and separate, typically representing counts or distinct categories. The possible outcomes can be listed or enumerated.

.Continuous Random Variables

A continuous random variable can take on an infinite number of values within a certain range. These values are not countable, and they represent measurements that can vary smoothly over a range.

Covariance measures the degree to which two variables change together. Correlation measures the strength and direction of a linear relationship between two variables. Unlike covariance, correlation is scaled between -1 and +1, making it easier to interpret.
